### The crime data used with this activity is a subset of dataset that was acquired from the FBI Uniform Crime Reporting program's "Offenses Known and Clearances by Arrest" database for the year in question, held at the National Archives of Criminal Justice Data. The data was compiled and analyzed by Gabriel Dance, Tom Meagher, and Emily Hopkins of The Marshall Project; the analysis was published as Crime in Context on 18 August 2016.

```python
# Render our plots inline
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt

# Make the graphs a bit prettier, and bigger
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (15, 5)
```

# 1.1 Reading data from a csv file

You can read data from a CSV file using the `read_csv` function. By default, it assumes that the fields are comma-separated.

We're going to be looking at some crime data for the Charlotte area.


```python
crime_df = pd.read_csv('clt_crime.csv')
```

```python
# Look at the first 5 rows
crime_df[:5]
```

You'll notice that this worked and displayed the data we wanted in table format.

# 1.2 Selecting a column

When you read a CSV, you get a kind of object called a `DataFrame`, which is made up of rows and columns. 

We can get the size of the data frame using `.shape` which will display a tuple with the number of rows and number of columns. 

```python
crime_df.shape
```

You get columns out of a DataFrame the same way you get elements out of a dictionary. Have you used dictionaries before?


Here's an example of selecting a column by its name and showing only the first 5 rows:

```python
crime_df['assaults'][:5]
```

Another way to display to the first 5 rows is use '.head(5)'

```python
crime_df.head(5)
```

You can also select a range of columns and rows to show

```python
# select the first five rows  and second tofourth column

crime_df[['agency_code','agency_jurisdiction','population']][:5]
```

```python
# another way to select the first five rows and second to fourth column
crime_df[crime_df.columns[1:4]][:5] 
```

# 1.3 Plotting a column

Pandas has tight integration with matplotlib. You can plot data directly from your DataFrame by simply adding `.plot()` to the end!


```python
crime_df['violent_crimes'].plot()
```

We can also plot two columns against each other

```python
crime_df.plot( x="report_year", y="violent_crimes")
```

If we are interested in visualizing multiple features we can plot more than one column just as easily. 
We can see that robbery incidents occur a lot more than homicides across all of the years.

```python
crime_df.plot(x="report_year", y=[ "homicides", "robberies"])
```
